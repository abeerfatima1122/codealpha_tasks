```python
#missing values treatment
df.isnull().sum()  #we donot do treatment for the target variable
```

```python
for i in ['Name','Availability','Brand','Description','Category',	'Review','Star','Date', 'Labels']:
   df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
from sklearn.impute import KNNImputer #it will take the avg of the nearest neighbour
impute=KNNImputer()
```

```python
for i in df.select_dtypes(include='number').columns:
  df[i]=impute.fit_transform(df[[i]])
```

```python
df.isnull().sum()
```

**Outliers Treatment**

**As the outlier is in target data so no need to do treatment**


```python
df.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt

value_counts = df['Availability'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(value_counts.index, value_counts.values, color='orange')
plt.title('Distribution of Availability')
plt.xlabel('Availability')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
value_counts = df['Brand'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(value_counts.index, value_counts.values, color='blue')
plt.title('Distribution of Brand')
plt.xlabel('Brand')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
value_counts = df['Category'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(value_counts.index, value_counts.values, color='green')
plt.title('Distribution of Category')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
value_counts = df['Star'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(value_counts.index, value_counts.values, color=['#FFD700', '#C0C0C0', '#FFA07A', '#90EE90', '#00BFFF'])
plt.title('Distribution of Star Ratings')
plt.xlabel('Star Rating')
plt.ylabel('Frequency')
plt.xticks(rotation=0, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
value_counts = df['Labels'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(value_counts.index, value_counts.values, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf'])
plt.title('Distribution of Labels')
plt.xlabel('Labels')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for column in df.select_dtypes(include='number').columns:
  plt.figure(figsize=(10, 6))
  sns.histplot(df[column], bins=20, color='skyblue')  # Use a unique color
  plt.title(f'Distribution of {column}')
  plt.xlabel(column)
  plt.ylabel('Frequency')
  plt.tight_layout()
  plt.show()
```

```python
import matplotlib.pyplot as plt
category_counts = df['Category'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Product Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
